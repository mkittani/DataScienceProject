```python
import pandas as pd
top_movies=pd.read_csv(r"C:\Users\2021\Downloads\imdb_top_1000.csv")
top_movies
```

```python
top_movies.dropna()
top_movies.dropna(axis="columns")
```

```python
top_genre=top_movies.Genre.value_counts().idxmax()
top_genre
```

```python
count_of_movies= pd.DataFrame(top_movies.Genre.value_counts())
count_of_movies
```

```python
import matplotlib.pyplot as pt
count_of_movies=count_of_movies.drop(count_of_movies.index[40:202])
count_of_movies.plot.bar()
pt.xlabel('genres')
pt.ylabel('count of movies/TV shows')
pt.show()
```

in the bar chart we know that the drama genre is the most popluar genre but since we have a lot of genres the bar chart
